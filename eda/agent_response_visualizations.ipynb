# Visualizations

## overall agreement vs disagreement

```python
def count_vals(df_inpt):
    overall_li = []
    for i in df_inpt.columns:
        for j in list(df_inpt[i]):
            overall_li.append(j)
            
    
    number_counts = Counter(overall_li)

    x_values = list(number_counts.keys())
    y_values = list(number_counts.values())

    return (x_values,y_values)
```

```python
biden = df_with_support[df_with_support["biden_or_trump"]==1]
biden = biden.drop(columns=["biden_or_trump"])
trump = df_with_support[df_with_support["biden_or_trump"]==-1]
trump = trump.drop(columns=["biden_or_trump"])
neutral = df_with_support[df_with_support["biden_or_trump"]==0]
neutral = neutral.drop(columns=["biden_or_trump"])
```

```python
plt.bar(count_vals(biden)[0], count_vals(biden)[1])
plt.xlabel("Level of Agreement")
plt.ylabel("Frequency")
plt.title("How often each result was outputted for Biden supporters")
```

```python
plt.bar(count_vals(trump)[0], count_vals(trump)[1])
plt.xlabel("Level of Agreement")
plt.ylabel("Frequency")
plt.title("How often each result was outputted for Trump supporters")
```

```python
plt.bar(count_vals(neutral)[0], count_vals(neutral)[1])
plt.xlabel("Level of Agreement")
plt.ylabel("Frequency")
plt.title("How often each result was outputted for neutral supporters")
```


- <b>Agents are WAY more likely to agree than disagree (code is probably skewed towards agreement).
- <b>It is interesting that the opinionated agents are more likely to be neutral, wheras the neutral agents' most frequent answer is weak agreement. This seems anti-intuitive, and requires further exploration.

## Jaccard similarity between different agents

```python
def jaccard_binary(list1,list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union
```

```python
agents_li = []
for i in list(biden.index):
    agents_li.append(i)
for i in list(neutral.index):
    agents_li.append(i)
for i in list(trump.index):
    agents_li.append(i)
```

```python
labels = agents_li
matrix = []

for i in agents_li:
    inner = []
    for j in agents_li:
        val = jaccard_binary((list(df.loc[i])),(list(df.loc[j])))
        inner.append(-val)
    matrix.append(inner)

plt.imshow(matrix)
plt.axhline(y=len(list(biden.index))+0.5, color='black', linewidth=2)
plt.axvline(x=len(list(biden.index))+0.5, color='black', linewidth=2)
plt.axhline(y=len(list(neutral.index))+ len(list(biden.index))+0.5, color='black', linewidth=2)
plt.axvline(x=len(list(neutral.index))+ len(list(biden.index))+ 0.5, color='black', linewidth=2)

plt.xticks(range(len(labels)), labels)
plt.yticks(range(len(labels)), labels)
plt.title(" Jaccard Simialirty, by candidate ")
```


#### agents have: 
    - have higher jaccard similaries with agents that support the same candidate as them 
    - lower jaccard similarities with agents that support a different candidate

# Overall conclusions

- Bias towards agreeemnt
    - It seems that the agent is skewed towards agreement. The agent only agrees or is neutral. It never disagrees. However, this might be due to the questions we chose. 
    - Next step: Ask the agent two different questions. Use opposite statements and see if the agent equally agrees with both. If yes, we know the agent is biased towards agreement, regardless of the question. If no, we know the agent is not biased. 
    
- Bias towards Biden
    - It seems that the agents lean towards the Democratic candidate. We ran our agents multiple times and did our best to produce agents that were not politcally biased. However, the best we could find is 6 Biden supporters, 3 Trump supporters, and one neutral supporter. 
    - We discussed this with our mentor, who mentioned that this is an issue with large language models as a whole. They are largely trained on data that is skewed towards more liberal values. Although we trained our models with a nearly equal split of datapoints from the Trump file and the Biden file, the model is skewed left.  
    
- Some topics are clearly more controversial than others. 
    - The topics have differing levels of agreement. 
    
- Agents that support the same candidate have similar opiniions. 
    - The Jaccard similarity plot shows us that agents support the same candidates generally have the same opinion on most of the questions. The Biden agents have a darker square with other Biden agents, and lighter sqaures with Trump agents. This shows that the agents seem to be able to replicate the thought pattern of actual voters.
